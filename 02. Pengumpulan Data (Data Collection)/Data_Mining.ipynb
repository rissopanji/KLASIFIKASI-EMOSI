```python
# Tentukan direktori tujuan
output_dir = './Dataset/NonLabel/Label_generate'

# Buat direktori tujuan jika belum ada
os.makedirs(output_dir, exist_ok=True)

# Tentukan jumlah data per bagian
chunk_size = 5000

# Bagi dataset menjadi beberapa bagian
for i, start in enumerate(range(0, len(data), chunk_size)):
    chunk = data.iloc[start:start + chunk_size]
    output_path = os.path.join(output_dir, f'data_{i + 1}.csv')
    chunk.to_csv(output_path, index=False)
    print(f'Bagian {i + 1} disimpan ke {output_path}')
```

### Pembuatan Fungsi

```python
import openai
from openai import AzureOpenAI
import pandas as pd

def create_prompt(text):
    role = "AI Linguistik"
    action = "menilai emosi dari tweet"
    step = "menganalisis tweet untuk menentukan emosi yang sesuai"
    context = "menentukan apakah emosi dari teks tersebut termasuk Neutral, Joy, Fear, Love, Sad, atau Anger"
    format_str = ("menjawab dengan salah satu dari enam emosi (Neutral, Joy, Fear, Love, Sad, atau Anger) "
                  "berdasarkan analisis emosi dari teks berikut:\n\n")

    prompt = (f"Anda adalah {role} yang {action}. {step}. {context}. {format_str}"
              f"{text}")
    return prompt

# Initialize Azure OpenAI client
client = AzureOpenAI(
    api_version="2023-05-15",
    azure_endpoint="https://chatbot-aic.openai.azure.com/",
    api_key="fd068a3036e34fe188a28392699ecc65",
)

# Function to get emotion of a text
def get_emotion(text, idx):
    prompt = create_prompt(text)
    print(f"{idx} - Text : {text}")
    try:
        response = client.chat.completions.create(
            model="aicdeploymodel",  # Ganti dengan nama deployment Anda
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt}
            ],
            max_tokens=100,
            temperature=0.3
        )
        emotion_response = response.choices[0].message.content
        emotions = ["Neutral", "Joy", "Fear", "Love", "Sad", "Anger"]
        for emotion in emotions:
            if emotion in emotion_response:
                print(f"Hasil Analisis : {emotion}\n")
                return emotion
    except Exception as e:
        # Handle any exception by skipping the text
        print(f"Error for text ke-{idx}: {text}, error: {e}")
        return "Skipped due to content policy"
```
