### 5. Dataset (Twitter Dataset)

---

Penelitian  : "Indonesian-Twitter-Emotion-Dataset"

Referensi   : https://www.kaggle.com/code/nikhil00/emotional-analysis-on-indonesian-tweets/input


Label       : "joy, fear, trust, anger, anticipation, disgust, sadness, surprise"

```python
import json

# Path ke file JSON
file_path = './Dataset/Label/5/emotion_id_tweets.json'

# Membaca file JSON baris per baris dan menggabungkannya menjadi satu objek JSON besar
data = []
with open(file_path, 'r') as file:
    for line in file:
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")

# Ekstraksi konten dan label
rows = []
for entry in data:
    content = entry['content']
    labels = entry['annotation']['labels']
    if labels:  # Memastikan bahwa daftar label tidak kosong
        label = labels[0]
    else:
        label = None  # Atau bisa dilewati dengan continue jika tidak ingin menyertakan entri tanpa label
    rows.append([content, label])

# Buat DataFrame
df = pd.DataFrame(rows, columns=['content', 'label'])

# Path output ke file CSV
csv_output_path = './Dataset/Label/5/dataset.csv'

# Simpan ke CSV
df.to_csv(csv_output_path, index=False)

print(f"Data telah disimpan ke {csv_output_path}")
print(df.head())
```

```python
dataset_kaggle = pd.read_csv("./Dataset/Label/5/dataset.csv", on_bad_lines="skip", delimiter=",")
dataset_kaggle.to_csv("./Dataset/Label/5/(5)Dataset_Kaggle.csv", sep="\t", index=False)
dataset_kaggle
```
