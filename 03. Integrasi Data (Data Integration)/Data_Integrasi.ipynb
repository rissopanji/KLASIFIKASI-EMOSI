### 5. Dataset (Twitter Dataset)

---
Kolom       : content, label (perubahaan)

Label       : "joy, fear, trust, anger, anticipation, disgust, sadness, surprise" (perubahan)

```python
dataset_kaggle = pd.read_csv(
    '../02. Pengumpulan Data (Data Collection)/Dataset/Label/5/(5)Dataset_Kaggle.csv',
      sep="\t"
)

dataset_kaggle
```

```python
dataset_kaggle = dataset_kaggle.rename(columns={'content': 'tweet'})
dataset_kaggle
```

Penjelasan :

1. Joy (Bliss, Contentment, Pride):
Kelas Emosi: Joy

2. Fear (Horror, Worry):
Kelas Emosi: Fear

3. Trust:
Kelas Emosi: Love (Karena Trust biasanya terkait dengan rasa kasih sayang atau kepercayaan yang tinggi)

4. Anger (Annoyance, Contempt, Hostility, Jealousy):
Kelas Emosi: Anger

5. Anticipation:
Kelas Emosi: Neutral (Anticipation tidak tercermin secara eksplisit dalam pohon emosi ini, sehingga dapat diklasifikasikan sebagai netral)

6. Disgust (termasuk dalam kategori Anger dengan Contempt):
Kelas Emosi: Anger

7. Sadness (Agony, Grief, Guilt, Loneliness):
Kelas Emosi: Sad

8. Surprise:
Kelas Emosi: Neutral (Surprise tidak tercermin secara eksplisit dalam pohon emosi ini, sehingga dapat diklasifikasikan sebagai netral)


Berikut adalah ringkasan pengelompokan berdasarkan gambar pohon emosi:

1. Neutral: Anticipation, Surprise

2. Joy: Joy

3. Fear: Fear

4. Love: Trust

5. Sad: Sadness

6. Anger: Anger, Disgust

```python
label_mapping = {
    'joy': 'Joy',
    'fear': 'Fear',
    'trust': 'Love',
    'anger': 'Anger',
    'anticipation': 'Neutral',
    'disgust': 'Anger',
    'sadness': 'Sad',
    'surprise': 'Neutral'
}

# Mengganti nilai pada kolom 'label'
dataset_kaggle['label'] = dataset_kaggle['label'].map(label_mapping)

dataset_kaggle
```

## C. Data Cleansing

---

Proses ini mencakup identifikasi dan penyelesaian masalah ketidakkonsistenan dalam data, seperti menghapus duplikasi, menangani nilai yang hilang dengan imputasi atau penghapusan, dan memperbaiki data yang bertentangan.

```python
dataset_integration = dataset_integration.dropna(subset=['tweet'])
```

```python
dataset_integration.drop_duplicates(subset=['tweet'], inplace=True)
```

```python
# Menentukan 6 label emosi
labels = ['Anger', 'Fear', 'Neutral', 'Sad', 'Joy', 'Love']

# Menghapus baris yang tidak memiliki label dari 6 emosi
dataset_integration = dataset_integration[dataset_integration['label'].isin(labels)]
```
